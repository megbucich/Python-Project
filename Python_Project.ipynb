```python
import matplotlib.pyplot as plt
import numpy as np
options = ['sin(2np.pi10x)' , 'e^(-x)*cos(5x)' , 'sinc(x)' , 'x^(3)+x']
chosen_functions = []

print('here are your function options: ' + '\n' )
for option in options:
  print(option)
while options:
  function = input('\nplease select as many functions as you want to see graphed. enter one at a time. When you want to quit, enter "done": ')
  if function == 'done':
    break
  else:
    options.remove(function)
    chosen_functions.append(function)

ha_min = int(input('\nbefore we present you your graphs, we need some more information. \nplease enter a beginning number for the start of the x axis: '))
ha_max = int(input('now, please enter a ending number for the x axis: '))
visual = input('would you like to see a separate graph for each funcation or one singular graph with every function on it? \nEnter either "separate" or "singular": ')
if visual == 'separate':
  x = np.linspace(ha_min, ha_max,100)
  fig = plt.figure(figsize=(20,10))
  while chosen_functions:
    if 'sinc(x)' in chosen_functions:
      ax1 = fig.add_subplot(2,2,1)
      ax1.plot(x, np.sinc(x))
      ax1.set(xlabel='x',ylabel='y',title='sinc(x)')
      chosen_functions.remove('sinc(x)')
    if 'e^(-x)*cos(5x)' in chosen_functions:
      ax2 = fig.add_subplot(2,2,2)
      ax2.plot(x,np.exp(-x)*np.cos(5*x) )
      ax2.set(xlabel='x',ylabel='y',title='e^(-x)*cos(5x)')
      chosen_functions.remove('e^(-x)*cos(5x)')
    if 'sin(2np.pi10x)' in chosen_functions:
      ax3 = fig.add_subplot(2,2,3)
      ax3.plot(x, np.sin(np.pi*2*10*x))
      ax3.set(xlabel = 'x' , ylabel = 'y', title = 'sin(2np.pi10x)')
      chosen_functions.remove('sin(2np.pi10x)')
    if 'x^(3)+x' in chosen_functions:
      ax4 = fig.add_subplot(2,2,4)
      ax4.plot(x, np.power(x,3)+x)
      ax4.set(xlabel = 'x' , ylabel = 'y', title = 'x^(3)+x')
      chosen_functions.remove('x^(3)+x')

elif visual == 'singular':
  x = np.linspace(ha_min, ha_max,100)
  fig = plt.figure(figsize=(10,5))
  if 'sinc(x)' in chosen_functions:
    plt.plot(x, np.sinc(x), label = 'sinc(x)', color = 'r')
    plt.legend()
  if 'e^(-x)*cos(5x)' in chosen_functions:
    plt.plot(x,np.exp(-x)*np.cos(5*x), label = 'e^(-x)*cos(5x)', color = 'g' )
    plt.legend()
  if 'sin(2np.pi10x)' in chosen_functions:
    plt.plot(x, np.sin(np.pi*2*10*x), label = 'sin(2np.pi10x)', color = 'b')
    plt.legend()
  if 'x^(3)+x' in chosen_functions:
    plt.plot(x,(np.power(x,3))+x, label = 'x^(3)+x', color = 'k')
    plt.legend()


```

-a list was created to name the four function options

-the code starts by giving the user their options of functions to choose from

-a while loop is used to ask them for their options and removes the option from the list because they are only allowed to pick four

-when they choose a function, it gets stored in a new list to keep track of the functions they want to see graphed.

-once the functions are chosen, the user is prompted to set their upper and lower limits and all of their graphs will have the same limits. This is done by using user input to collect their numbers. the numbers are used to set the x axis using linspace

-after they select their options, they get to choose weather they want one graph per function or each function on one graph through user inputs. following this user input, there is an if and elif statement for both options

-if they picked separate graphs, the while loops will go through the chosen list of functions and go through four more if statements. The if statements include information about each individual fucntion and sets up the graph with proper code. They are if statements because they could have more than one of the if statements be true. it will plot all of these on the same figure, but each function will have its own graph.

if they picked singular graph, then the elif statement with go through the sub-if statements to see which functions they picked. This will plot all the functions on one graph and has different elements for each function to better distinguish between each one.
